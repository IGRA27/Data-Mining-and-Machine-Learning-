```python
#Importo todas las librerias necesarias para analizar el CSV, Supervisado
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('exportaciones_eeuu2023.csv')
```

```python
print(df.head())
```

```python
df.info()
```

```python
# Verificar si hay valores nulos en el DataFrame
print(df.isnull().sum())
```

```python
df.describe().transpose()
```

```python
# Eliminar filas con valores nulos
df = df.dropna()
```

```python
# Eliminar duplicados en el DataFrame
df = df.drop_duplicates()
```

```python
# Ejemplo de análisis y visualización
# Graficar el total de exportaciones por año
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x='Año', data=df)
plt.title('Total de Exportaciones por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Exportaciones')
plt.xticks(rotation=45)
plt.show()
```

```python
#ANALISIS DEL GRAFICO:
```

Al observar el gráfico, podemos ver que las exportaciones fueron bastante estables desde el inicio del siglo hasta 2006, con una ligera caída en el 2001-2002, probablemente debido a las repercusiones económicas del ataque a las Torres Gemelas.

Luego, hubo una disminución significativa en 2007-2008, que podría estar relacionada con la crisis financiera global.

Vemos un aumento notable en las exportaciones a partir del 2009, y este crecimiento se mantiene hasta 2018. Este crecimiento podría estar asociado con una combinación de factores, como la recuperación económica después de la crisis del 2008, y posiblemente las políticas comerciales favorables entre Ecuador y los Estados Unidos.

La caída en las exportaciones en 2019-2020 se puede atribuir a la pandemia de COVID-19, que interrumpió el comercio y la economía global.

En 2021-2022, hay un notable repunte en las exportaciones, que podría ser el resultado de la recuperación económica a medida que las restricciones de la pandemia se suavizan y el comercio se reanuda. Sin embargo, este es solo un análisis superficial y la causa precisa de estas fluctuaciones en las exportaciones requeriría una investigación más detallada y probablemente la combinación de datos económicos, políticos y comerciales más amplios.

```python
sns.scatterplot(x='Export FOB (miles)', y='Toneladas', data=df)
plt.title('Comparación de Export FOB (miles) VS. Toneladas')
plt.xlabel('Export FOB (miles)')
plt.ylabel('Toneladas')
plt.show
```

```python
#Analisis del grafico:
```

Puedo notar y podemos notar que hay una alta concentración de datos cerca del origen (0,0), lo que sugiere que una gran cantidad de las exportaciones son de bajo peso y valor. Es posible que estos sean productos de alto valor por unidad de peso, como ciertos productos de consumo.

Además, mencionas que los datos se dispersan más a medida que nos movemos hacia la derecha. Esto podría indicar que hay menos exportaciones de alto peso y/o valor, lo cual tendría sentido si consideramos que estos envíos serían más costosos y por lo tanto menos frecuentes.

En términos de la relación entre Export FOB (miles) y Toneladas, parece que hay cierta correlación positiva pero no lineal. En los rangos que mencionas, cuando Export FOB (miles) aumenta, Toneladas también aumenta, pero no al mismo ritmo. Esto podría indicar que a medida que el valor de las exportaciones aumenta, también lo hace el peso, pero a una tasa decreciente. Esta podría ser una señal de que las exportaciones más valiosas contienen una mezcla de productos, algunos de los cuales son menos pesados pero de mayor valor.
