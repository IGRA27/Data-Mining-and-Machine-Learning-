(g) Create an array of 10 random integers between 1 and 100, and then calculate the mean and standard deviation of the array.

```r
set.seed(123)  # Para reproducibilidad

# Crear vector de 10 enteros aleatorios entre 1 y 100
vector <- sample(1:100, 10, replace = TRUE)

# Calcular media y desviación estándar del vector
media <- mean(vector)
desviacion_estandar <- sd(vector)

# Imprimir los resultados
cat("Vector:", vector, "\n")
cat("Media:", media, "\n")
cat("Desviación estándar:", desviacion_estandar, "\n")
```

(h) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column.

```r
set.seed(123)  # Para reproducibilidad

# Crear matriz de 3 filas y 4 columnas con valores enteros aleatorios
matriz <- matrix(sample(1:100, 12, replace = TRUE), nrow = 3)

# Calcular la suma de cada fila y columna
suma_filas <- rowSums(matriz)
suma_columnas <- colSums(matriz)

# Imprimir los resultados
cat("Matriz:\n")
print(matriz)

cat("Suma de cada fila:\n")
print(suma_filas)

cat("Suma de cada columna:\n")
print(suma_columnas)
```

(i) Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than or equal to 30.

```r
# Crear DataFrame
df <- data.frame(
  Name = c("John", "Alice", "Michael", "Emily", "David"),
  Age = c(25, 32, 28, 36, 41),
  City = c("New York", "London", "Paris", "Sydney", "Tokyo")
)

# Filtrar el DataFrame por edad mayor o igual a 30
df_filtrado <- subset(df, Age >= 30)

# Imprimir el DataFrame filtrado
print(df_filtrado)
```

(k) Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame.

```r
library(readr)

# Cargar archivo CSV en un DataFrame
df <- read_csv("/kaggle/input/csvdata/Network.csv")  # Reemplaza "tu_archivo.csv" con el nombre o ruta de tu archivo CSV

# Calcular la media, mediana y moda de una columna
columna <- df$IssuerId  

media <- mean(columna)
mediana <- median(columna)
moda <- names(table(columna))[table(columna) == max(table(columna))]

# Imprimir las estadísticas calculadas
cat("Media: ", media, "\n")
cat("Mediana: ", mediana, "\n")
cat("Moda: ", moda, "\n")
```

(l) Create a scatter plot of random x and y values between 1 and 100.

```r
# Generar valores aleatorios de x e y
set.seed(123)  # Para reproducibilidad
x <- runif(50, min = 1, max = 100)
y <- runif(50, min = 1, max = 100)

# Crear el gráfico de dispersión
plot(x, y, main = "Gráfico de Dispersión", xlab = "X", ylab = "Y")

# Agregar título y etiquetas de ejes
title(main = "Gráfico de Dispersión")
xlabel <- expression(X)
ylabel <- expression(Y)
axis(side = 1, label = xlabel)
axis(side = 2, label = ylabel)

# Mostrar el gráfico de dispersión
```
