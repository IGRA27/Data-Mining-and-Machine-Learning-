(a) Create an array of 10 random integers between 1 and 100, and then calculate the mean and standard deviation of the array.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
#Create an array of 10 random integers between 1 and 100
a=[25,16,66,55,99,42,57,91,85,8]
print(len(a),a)
print(type(a))
#then calculate the mean and standard deviation of the array.
print(np.mean(a))
print(np.std(a))
```

```python
b=np.random.randint(1, high=101, size=10)
print(len(b),b)
print(type(b))
print(b.mean())
print(np.std(b))
```

(c) Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than or equal to 30.


```python
import pandas as pd

# Create DataFrame
data = {
    'Name': ['John', 'Alice', 'Michael', 'Emily', 'David'],
    'Age': [25, 32, 28, 36, 41],
    'City': ['New York', 'London', 'Paris', 'Sydney', 'Tokyo']
}

df = pd.DataFrame(data)

# Filter the DataFrame
filtered_df = df[df['Age'] >= 30]

# Display the filtered DataFrame
print(filtered_df)
```

(d) Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame.

```python
import pandas as pd

# Load CSV file into DataFrame
df = pd.read_csv('Network.csv')  

# Calculate mean, median, and mode of a column
column_name = 'IssuerId'  

column = df[column_name]

mean = column.mean()
median = column.median()
mode = column.mode().values[0]

# Print the calculated statistics
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

(e) Create a scatter plot of random x and y values between 1 and 100.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random x and y values
np.random.seed(123)  # reprod.
x = np.random.randint(1, 101, size=50)
y = np.random.randint(1, 101, size=50)

# Create scatter plot
plt.scatter(x, y)

# Set plot title and labels
plt.title('Scatter Plot of Random Values')
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.show()
```

(f) Load a CSV file into a DataFrame and then create a line chart of one of the columns in the DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load CSV file into DataFrame
df = pd.read_csv('ServiceArea.csv')  
# Create line chart of a column
column_name = 'County' 

column = df[column_name]

# Create line chart
plt.plot(column)

# Set plot title and labels
plt.title('Line Chart of ' + column_name)
plt.xlabel('Index')
plt.ylabel(column_name)

# Display the chart
plt.show()
```
